**# Import necessary libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

**# Load the dataset**

```python
import pandas as pd
data = pd.read_csv("/kaggle/input/titanic-dataset/train.csv")
```

**# Display first few rows**

```python
data.head()
```

**# Basic Information about the dataset**

```python
data.info()
```

**# Statistical Summary of the dataset**

```python
data.describe()
```

**# Check for missing values**

```python
data.isnull().sum()
```

**# Value counts for categorical columns**

```python
for col in data.select_dtypes(include='object').columns:
    print(f"\nValue counts for {col}:")
    print(data[col].value_counts())
```

# Visual Explorations

# Histograms for numerical columns

```python
data.hist(figsize=(15, 10), bins=30)
plt.suptitle('Histograms of Numerical Columns')
plt.show()
```

# Boxplots to check for outliers


```python
num_cols = data.select_dtypes(include=['float64', 'int64']).columns
for col in num_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x=data[col])
    plt.title(f'Boxplot of {col}')
    plt.show()
```

# Scatterplots (between first two numerical columns if available)

```python
if len(num_cols) >= 2:
    plt.figure(figsize=(6, 4))
    sns.scatterplot(x=data[num_cols[0]], y=data[num_cols[1]])
    plt.title(f'Scatterplot between {num_cols[0]} and {num_cols[1]}')
    plt.show()
```

# Pairplot

```python
sns.pairplot(data.select_dtypes(include=['float64', 'int64']).dropna().sample(100))
plt.suptitle('Pairplot of Numerical Features', y=1.02)
plt.show()
```
